## Understanding the Output

Key metrics in the log:

- **train_loss**: Cross-entropy loss on training data (nats)
- **val_loss**: Cross-entropy on validation (nats)
- **val_bpb**: Bits per byte (this is what the leaderboard ranks by)
- **final_int8_zlib_roundtrip**: The official score after compression

**Good BPB values**:
- Baseline (8xH100): 1.2244
- SOTA (8xH100): 1.0810
- Your Colab run: ~1.5-2.0 (normal, you're on 1 GPU with less training)

Colab is for **learning**, not competing. For real submissions, rent 8xH100 GPUs.
